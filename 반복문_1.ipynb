# 반복문
반복문은 자료구조의 데이터 수만큼 코드를 반복실행 합니다.  
자료구조에 있는 "데이터를 하나씩 바인딩하고 들여쓰기된 코드를 실행" 과정을 반복하는 겁니다. 
  
## 반복문 - 기본형
`for 변수 in 자료구조 : ` 형태로 코드를 작성합니다. 코드 끝에 콜론 `:` 있음에 주의하세요. 분기문과 같이 콜론 다음에는 들여쓰기와 함께 영향을 받는 코드를 작성합니다.   
-> 들여쓰기된 한 덩어리의 코드 : 키워드(for)의 영향을 받는 구간  
-> 분기문에서도 들여쓰기 된 코드는 분기문(if)의 영향을 받는 구간

###### 반복문으로 코드를 정리하기

````python
for 변수 in 리스트: # 리스트는 순서가 있는 자료구조 -> 따라서, for 문도 index[0] 자료부터 출력
  print(변수)
````

````python
리스트 = ["철수", "영희", "희철"] # 'list'는 python keyword이기 때문에 변수명으로 설정하지 않기

for 승객 in 리스트:
  print(f"{승객} 탑승 예정")  
````

###### 연습문제 - 2
리스트에서 앞의 두 개의 값을 다음과 같이 출력하라.  
```
가
나
```

````python
# 슬라이싱 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
````

````python
# indexing : 하나의 데이터를 선택. 해당 위치에 있는 값을 가져온다
# slicing : 하나 이상의 데이터를 선택. 꺼내온 결과를 다시 리스트에 담아줌(문자열을 슬라이싱하면 문자열, 리스트를 슬라이싱하면 리스트 타입, 시리즈를 슬라이싱하면 시리즈 타입)
for i in 리스트[:2]:     # 리스트[ : ] = ['가','나','다']  /   리스트[ : : 2] = ['가','다']  / 리스트[ : : -1] = ['다','나','가']
  print(i)
````

````python
# 분기문을 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
for 변수 in 리스트:
  if 변수 != '다':  # 분기문은 조건이 참인 경우에만 들여쓰기된 코드를 실행
    print(변수)
````

````python
num = 0   # 초기화
for i in 리스트:
  num = num + 1
  if num < 3:  # num은 지역 변수
    print(i)
print(i)  # for문이 끝나면 리스트의 마지막 값을 계속 가리키고 있다
````

###### 연습문제 - 4
`xls` 파일만을 출력하라

````python
리스트 = ["ab.xls", "a.hwp", "b.exe", "bbbbb.xls"]
for 변수 in 리스트:  # 변수는 리스트의 원소를 하나씩 바인딩 
  if 변수[-3:] == 'xls':
    print(변수)
  # 참고 : 변수[-3:0] ... 역방향으로 메모리에 있는 데이터를 가져올 수는 없다
````

````python
for 변수 in 리스트:
  if 변수.endswith("xls"):
    print(변수)
````

````python
for 변수 in 리스트:
  if ".xls" in 변수:
    print(변수)
````

````python
# 결과를 리스트에 저장
xls_list = []  # 초기화 : 내가 사용할 data type과 자료구조를 정의해야 python interpreter가 제대로 작동
for 변수 in 리스트:
  if 변수.endswith("xls"):
    xls_list.append(변수)
print(xls_list)
````

###### 연습문제 - 7
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.   
```
1999년 성적: 수 
2000년 성적: 양 
2001년 성적: 미 
2002년 성적: 수 
```

````python
my_grade = ["1999-수", "2000-양", "2001-미", "2002-수"]
for i in my_grade:
  연도 = i[:4]
  성적 = i[-1]
  print(f"{연도}년 성적 : {성적}")
````

````python
my_grade = ["1999-수", "2000-양", "2001-미", "2002-수"]
for i in my_grade:
  b = i.split('-')
  print(f"{b[0]}년 성적 : {b[1]}")
````
